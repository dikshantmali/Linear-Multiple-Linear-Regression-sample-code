```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
df = pd.DataFrame(boston.data,columns = boston.feature_names )
df
```

```python
df['price'] = boston.target
```

```python
df
```

```python
plt.hist(df['price'],bins = 10)
```

```python
df['price'].mean()
```

```python
lstat = df.LSTAT
price = df.price
```

```python
lstat = np.array(lstat).reshape(-1,1)
price = np.array(price).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(lstat,price,test_size = 0.1,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
rmse
```

```python
R2 = round(lr.score(x_test,y_test),3)
```

```python
R2
```

<h1>Now we will do data visualixation</h1>

```python
space = np.linspace(min(lstat),max(lstat)).reshape(-1,1)
plt.scatter(lstat,price)
plt.plot(space,lr.predict(space),color='black',linewidth=5)
plt.xlabel("Number of LSTAT")
plt.ylabel("Price Of Houses")
plt.title("LSTAT vs Price")
plt.show()
```
